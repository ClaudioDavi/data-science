
# Projeto K Means Clustering

### Apresentação

Neste projeto, tentaremos utilizar KMeans para tentar classificar universidades Americanas em dois grupos distintos: Privadas e públicas.

Geralmente quando utilizamos o K Means nós não temos os labels finais já e estamos tentando identificá-los. Porém, como esses dados já oferecem eles pra gente, vamos utilizá-los para verificar se o nosso algortimo acertou nas classificações.

### Objetivo

Separar as universidades em duas categorias, públicas e privadas.

### Significado das colunas

Nossos dados contém 777 entradas sobre universidades americanas, cada uma delas composta de 18 colunas, sendo elas:

* 'Private' - Um indicador composto por 'Yes' ou 'NO' caso a universidade é privada.

* 'Top25perc' - Novos estudantes dos 25% melhores do ensimo médio (High School)
* 'F.Undergrad' - Numero de estudantes em tempo integral
* 'P.Undergrad' - Número de estudantes em tempo parcial
* 'Outstate' - Mensalidade para fora do estado.
* 'Room.Board' - Preço de manutenção das salas
* 'Books' - Preço estimado dos livros
* 'Personal' Valor estimado gasto com a staff
* 'PhD' Porcentagem de pessoal com PhD
* Terminal Pct. of faculty with terminal degree
* S.F.Ratio Student/faculty ratio
* perc.alumni Pct. alumni who donate
* Expend Instructional expenditure per student
* Grad.Rate Graduation rate

## K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label.**

```python
kmeans.fit(college_data.drop('Private', axis=1))
```

** What are the cluster center vectors?**

```python
kmeans.cluster_centers_
```

## Evaluation
That is not a good idea, to evaluate. Because we are using an unsupervised method. However, we know if a school is private or not, we can compare results to see how well we did

```python
def cluster(column):
    return 1 if column == 'Yes' else 0

college_data['Cluster'] = college_data['Private'].apply(cluster)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(college_data['Cluster'], kmeans.labels_))
print(classification_report(college_data['Cluster'], kmeans.labels_))
```
